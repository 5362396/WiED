```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('wine.csv')

# Data for plotting
t = np.arange(0.0, 178.0, 1.0)
s = 1 + np.sin(2 * np.pi * df['Acl'])

fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(t, s, color='m', linewidth=3)

ax.set(xlabel='Time', ylabel='Pulse amplitude',
       title='Fake pulse using one column of data from wine.csv')
ax.grid(color='c', linestyle=':')

# fig.savefig("test.png")
plt.show()
```
