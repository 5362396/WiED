```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('ToxinsbaseClass.csv')
all_data = df.select_dtypes(['float64']).iloc[:, 3:8]
labels = all_data.columns

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 8))

# rectangular box plot
bplot1 = axes[0].boxplot(all_data,
                         vert=True,  # vertical box alignment
                         patch_artist=True,  # fill with color
                         labels=labels)  # will be used to label x-ticks
axes[0].set_title('Rectangular', color='royalblue', fontsize=20)

# notch shape box plot
bplot2 = axes[1].boxplot(all_data,
                         notch=True,  # notch shape
                         vert=True,  # vertical box alignment
                         patch_artist=True,  # fill with color
                         labels=labels)  # will be used to label x-ticks
axes[1].set_title('Notched', color='royalblue', fontsize=20)

# fill with colors
colors = ['orchid', 'lightseagreen', 'gold', 'palegreen', 'lavender']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

# adding horizontal grid lines
for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xlabel('Five toxinsbase samples', color='tomato', fontsize=12)
    ax.set_ylabel('Values given', color='burlywood', fontsize=15)
    for tick in ax.get_xticklabels():
        tick.set_rotation(55)

plt.show()
```
