```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Fixing random state for reproducibility
np.random.seed(8)

fig, ax = plt.subplots(figsize=(12, 12))

# Example data
df = pd.read_csv('houseprice.csv')
y_pos = np.arange(11)
performance = df['LotArea'][1:12]
error = df['LotFrontage'][1:12]

ax.barh(y_pos, performance, xerr=error, align='center', color='orange')
ax.set_yticks(y_pos, labels=df['Neighborhood'][1:12])
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Lot area', fontsize=15, fontweight='ultralight', color='r')
ax.set_title('Lot areas in different neighborhoods', fontsize=20, fontweight='bold', color='b')
ax.tick_params(labelcolor='g')

plt.show()
```
