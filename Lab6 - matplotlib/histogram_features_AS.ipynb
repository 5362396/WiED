```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('ToxinsbaseClass.csv')

# example data
mu = 20  # mean of distribution
sigma = 5  # standard deviation of distribution
x = df['T_Epi ']

num_bins = 50

fig, ax = plt.subplots(figsize=(12, 8))

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=1, facecolor='b', alpha=0.75, edgecolor='g', linewidth=3)

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, y, '-.', color='r', linewidth=3)
ax.yaxis.grid(True)
ax.set_xlabel('Quantity', fontsize=13, color='magenta')
ax.set_ylabel('Calculated Probability', fontsize=13, color='y')
ax.set_title(r'Histogram of T_Epi from ToxinsbaseClass: $\mu=20$, $\sigma=5$', fontsize=15, fontweight='bold', color='darkorange')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```
