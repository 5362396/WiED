```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import pandas as pd

df = pd.read_csv('wine.csv')

fig = plt.figure(tight_layout=True, figsize=(8, 8))
gs = gridspec.GridSpec(3, 3)

ax = fig.add_subplot(gs[:2, :])
ax.plot(df['Proline'], color='orange', linestyle='--')
ax.set_ylabel('Proline', fontsize=15, color='g')
ax.set_xlabel('Sample ID', fontsize=15, color='g')

for i in range(3):
    ax = fig.add_subplot(gs[2, i])
    ax.plot(df.iloc[:, i:i+1], color='maroon')
    ax.set_ylabel(df.columns[i], color='teal', fontsize=11)
    ax.set_xlabel('Sample ID', color='teal', fontsize=11)
    for tick in ax.get_xticklabels():
        tick.set_rotation(27 * i)
fig.align_labels()  # same as fig.align_xlabels(); fig.align_ylabels()

plt.show()
```
